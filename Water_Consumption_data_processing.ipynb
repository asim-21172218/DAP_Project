```python
from pprint import pprint
import urllib.request
import json
import pandas as pd
import pymongo


##opening and reading JSON FILE

with open("Water Consumption And Cost (2013 - Feb 2022).json", 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data['data'], columns=column_name)

df.head()
```

```python
#Finding columns with constant values

constant_column = [column for column in df.columns if df[column].nunique() == 1]

constant_column
```

```python
df.drop(["created_meta","updated_meta"], axis=1, inplace=True)
```

```python
date_na=df[df["Service Start Date"].isna()]

df = df.drop(date_na.index)
```

```python
df["Meter AMR"].unique()
```

```python
df["Revenue Month"] = pd.to_datetime(df["Revenue Month"], format='%Y-%m-%d')
```

```python
# filtering data as we are only considering data from 2013 and 2020

df = df.loc[(df["Revenue Month"] >'2012-12-31') | (df["Revenue Month"] < '2021-01-01')]

len(df)
```

```python
#Water&Sewer Charges,Other Charges,Estimated,Bill Analyzed,TDS #,EDP,RC Code,Location,Meter AMR,Funding Source,sid,id
clean_data = df[["Development Name","Borough","Account Name","Location","UMIS BILL ID","Revenue Month","Service Start Date","Service End Date","# days","Meter Number","Consumption (HCF)","Rate Class"]]
```

```python
clean_data.head()
```

```python
na_counts = clean_data.isnull().sum()
na_counts
```

```python
clean_data = clean_data.dropna(how='any')
na_counts = clean_data.isnull().sum()
na_counts
```
