```python
import numpy as np
import pandas as pd

path = "./nyc_temp_1869_2021.csv"

with open(path, 'r') as file:
    weather_data = pd.read_csv(path)
    
```

```python
na_counts = weather_data.isnull().sum()
na_counts
```

```python
weather_data
```

```python
weather_data["MM/DD/YYYY"] = pd.to_datetime(weather_data["MM/DD/YYYY"], format='%Y-%m-%d')

weather_data.dtypes
weather_data

weather_data.rename(columns={"MM/DD/YYYY": "DATE"}, inplace=True)
```

```python
weather_data
```

```python
weather_data.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
weather_data
```

```python
weather_data = weather_data.loc[(weather_data["DATE"] >'2012-12-31') & (weather_data["DATE"] < '2021-01-01')]
```

```python
weather_data

print(weather_data.shape)
```

```python
#dataframe conversion to list of dict
data_document=weather_data.to_dict('records')

from pymongo import MongoClient
#mongodb database connection using ip destination and port number

client=MongoClient("mongodb://%s:%s@127.0.0.1" % ("dap", "dap"))
#creating database named "DAP"
db=client["Database"]
#creating a table name "newyork_weather in DAP database"
collection=db["newyork_weather"]
collection.insert_many(data_document)
```

```python
#web scraping using python
#importing required packages
from bs4 import BeautifulSoup as bs
import requests

response=requests.get("https://www.wunderground.com/weather/us/ny/new-york-city")
#instantiating beautifulsoup class object
obj=bs(response.content)
#typecasting html tree structure to python interpretable
todays_temp=obj.find("div",{"class":"hi-lo"})
#print(todays_temp)

from datetime import date
today=date.today()
print(today.strftime("%B %d, %Y")+"\t Location: Newyork city, USA")
#extracting nested data content
temp_max=todays_temp.find("span",{"class":"hi"}).text
temp_min=todays_temp.find("span",{"class":"lo"}).text

print(f"Todays Highest Temperature is {temp_max} Fahrenheit")
print(f"Todays Highest Temperature is {temp_min} Fahrenheit")
```
